```python
def calculateUNIGAU(x, mean, std, type):                    
    exponent = exp(-((x - mean) ** 2 / (2 * std ** 2)))
    probability = (1 / (sqrt(2 * pi) * std)) * exponent
    if type == "w1" :
        return probability * 0.5
    elif type == "w2":
        return probability * 0.5
    else:
        return probability * 0.0

```

```python
def calculateUNIGAUMatrix(dataset, calculateMean_list, calculateSTD_list, type):
    columnList = ["x1", "x2", "x3"]
    UNIGAUMatrix = []
    for i in range(len(dataset)):
        for j in range(3):
            mean = calculateMean_list[j]
            stdev = calculateSTD_list[j]
            columnName = columnList[j]
            num = calculateUNIGAU(dataset[columnName][i], mean, stdev, type)
            UNIGAUMatrix.append(num)
    UNIGAUMatrix = np.array(UNIGAUMatrix)
    matrix = np.reshape(UNIGAUMatrix, (len(dataset), 3))
    return matrix

```

```python
def calculateMean(dataset):             #Mean
    columnList = ["x1", "x2", "x3"]
    calculateMean_list = []
    for i in range(3):
        columnName = columnList[i]
        mean = dataset[columnName].mean()
        calculateMean_list.append(mean)
    return calculateMean_list
```

```python
def calculateSTD(dataset):              #SD
    columnList = ["x1", "x2", "x3"] 
    calculateSTD_list = []
    for i in range(3):
        columnName = columnList[i]
        std = dataset[columnName].std()
        calculateSTD_list.append(std)
    return calculateSTD_list
```

```python
def comp_prob(col_w1, col_w2, col_w3, test):
    count = 0
    for j in range(len(col_w1)):
        if col_w1[j] > col_w2[j]:
            if col_w1[j] > col_w3[j]:
                label = "w1"
        elif col_w2[j] > col_w1[j]:
            if col_w2[j] > col_w3[j]:
                label = "w2" 
        else:
            label = "w3"
        if label != test["type"][j]:
            count = count + 1
    calculateError = (count / len(col_w1)) * 100
    return calculateError
```

```python
def calcA(calculateUNIGAUMatrix_w1, calculateUNIGAUMatrix_w2, calculateUNIGAUMatrix_w3, test):
    col1_w1 = [sub[0] for sub in calculateUNIGAUMatrix_w1]
    col1_w2 = [sub[0] for sub in calculateUNIGAUMatrix_w2]
    col1_w3 = [sub[0] for sub in calculateUNIGAUMatrix_w3]
    calculateError = comp_prob(col1_w1, col1_w2, col1_w3, test)
    print("Part A Results:" + str(calculateError) + "%")
```

```python
def calcB(calculateUNIGAUMatrix_w1, calculateUNIGAUMatrix_w2, calculateUNIGAUMatrix_w3, test):
    col_w1 = []
    col_w2 = []
    col_w3 = []
    for i in range(len(calculateUNIGAUMatrix_w1)):
        col_w1.append(calculateUNIGAUMatrix_w1[i][0] * calculateUNIGAUMatrix_w1[i][1])
        col_w2.append(calculateUNIGAUMatrix_w2[i][0] * calculateUNIGAUMatrix_w2[i][1])
        col_w3.append(calculateUNIGAUMatrix_w3[i][0] * calculateUNIGAUMatrix_w3[i][1])
    calculateError = comp_prob(col_w1, col_w2, col_w3, test)
    print("Part B Results" + str(calculateError) + "%")
```

```python
def calcC(calculateUNIGAUMatrix_w1, calculateUNIGAUMatrix_w2, calculateUNIGAUMatrix_w3, test):
    col_w1 = []
    col_w2 = []
    col_w3 = []
    for i in range(len(calculateUNIGAUMatrix_w1)):
        col_w1.append(calculateUNIGAUMatrix_w1[i][0] * calculateUNIGAUMatrix_w1[i][1] * calculateUNIGAUMatrix_w1[i][2])
        col_w2.append(calculateUNIGAUMatrix_w2[i][0] * calculateUNIGAUMatrix_w2[i][1] * calculateUNIGAUMatrix_w2[i][2])
        col_w3.append(calculateUNIGAUMatrix_w3[i][0] * calculateUNIGAUMatrix_w3[i][1] * calculateUNIGAUMatrix_w3[i][2])
    calculateError = comp_prob(col_w1, col_w2, col_w3, test)
    print("Part C Results" + str(calculateError) + "%")
```
